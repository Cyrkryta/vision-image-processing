```python
# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter

# Reading image
img = cv2.imread("lenna.jpg")
plt.imshow(img)
plt.axis("off")
```
