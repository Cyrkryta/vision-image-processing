```python
# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter, gaussian_laplace, sobel

# Reading image
img = cv2.imread("lenna.jpg", 0)
img = img / img.max()
plt.imshow(img, cmap='gray')
plt.axis("off")
```

```python
'''
    Gaussian filtering
'''

#sigmas = np.array([1, 2, 4, 8])
# def gauss_filter(sigma):
#     x = img.shape[0]
#     y = img.shape[1]
#     kernel = 1/(2 * np.pi * sigma**2)
#     exp = np.exp(-(x**2 + y**2)/(2 * sigma**2))
#     return kernel * exp

# for sigma in sigmas:

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

results = []

#
blurred_img = gaussian_filter(img, sigma=2)

for sigma in [1, 2, 4, 8]:
    results.append(gaussian_filter(img, sigma=sigma))
    
axs[0, 0].imshow(results[0], cmap="gray")
axs[0, 0].axis("off")
axs[0, 0].set_title("sigma = 1")

axs[0, 1].imshow(results[1], cmap="gray")
axs[0, 1].axis("off")
axs[0, 1].set_title("sigma = 2")

axs[1, 0].imshow(results[2], cmap="gray")
axs[1, 0].axis("off")
axs[1, 0].set_title("sigma = 4")

axs[1, 1].imshow(results[3], cmap="gray")
axs[1, 1].axis("off")
axs[1, 1].set_title("sigma = 8")
```

```python
'''
    Gradient Magnitude Computation - y
'''
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

results = []

for sigma in [1, 2, 4, 8]:
    results.append(gaussian_filter(blurred_img, sigma=sigma, order=[0,1]))
    
axs[0, 0].imshow(results[0], cmap="gray")
axs[0, 0].axis("off")
axs[0, 0].set_title("sigma = 1")

axs[0, 1].imshow(results[1], cmap="gray")
axs[0, 1].axis("off")
axs[0, 1].set_title("sigma = 2")

axs[1, 0].imshow(results[2], cmap="gray")
axs[1, 0].axis("off")
axs[1, 0].set_title("sigma = 4")

axs[1, 1].imshow(results[3], cmap="gray")
axs[1, 1].axis("off")
axs[1, 1].set_title("sigma = 8")
```

```python
'''
    Gradient Magnitude Computation - x
'''
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

results = []

for sigma in [1, 2, 4, 8]:
    results.append(gaussian_filter(blurred_img, sigma=sigma, order=[1,0]))
    
axs[0, 0].imshow(results[0], cmap="gray")
axs[0, 0].axis("off")
axs[0, 0].set_title("sigma = 1")

axs[0, 1].imshow(results[1], cmap="gray")
axs[0, 1].axis("off")
axs[0, 1].set_title("sigma = 2")

axs[1, 0].imshow(results[2], cmap="gray")
axs[1, 0].axis("off")
axs[1, 0].set_title("sigma = 4")

axs[1, 1].imshow(results[3], cmap="gray")
axs[1, 1].axis("off")
axs[1, 1].set_title("sigma = 8")
```

```python
'''
    Test gradient magnitude
'''
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

results = []

for sigma in [1, 2, 4, 8]:
    squared_x = gaussian_filter(blurred_img, sigma=sigma, order=[1,0]) ** 2
    squared_y = gaussian_filter(blurred_img, sigma=sigma, order=[0,1]) ** 2
    sqrt_xy = np.sqrt(squared_x + squared_y)
    results.append(sqrt_xy)
    
axs[0, 0].imshow(results[0], cmap="gray")
axs[0, 0].axis("off")
axs[0, 0].set_title("sigma = 1")

axs[0, 1].imshow(results[1], cmap="gray")
axs[0, 1].axis("off")
axs[0, 1].set_title("sigma = 2")

axs[1, 0].imshow(results[2], cmap="gray")
axs[1, 0].axis("off")
axs[1, 0].set_title("sigma = 4")

axs[1, 1].imshow(results[3], cmap="gray")
axs[1, 1].axis("off")
axs[1, 1].set_title("sigma = 8")
```
