```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
from skimage.filters import threshold_otsu
```

```python
pic1 = imread('C:/Users/renia/Documents/IT/VIP/VIP5/camera.png')
pic2 = imread('C:/Users/renia/Documents/IT/VIP/VIP5/coins.png')
pic3 = imread('C:/Users/renia/Documents/IT/VIP/VIP5/page.png')
pic4 = imread('C:/Users/renia/Documents/IT/VIP/VIP5/rocksample.png')
images = [pic1, pic2, pic3, pic4]

#display of original images
fig, ax = plt.subplots(1, 4, figsize=(15, 3))
for i, image in enumerate(images):
    # Display original image
    ax[i].imshow(image, cmap='gray')
    ax[i].set_title(f'pic{i+1} Original')
    
plt.tight_layout()
plt.show()

#Otsu's
fig, ax = plt.subplots(1, 4, figsize=(12, 4))

for i, image in enumerate(images):
    # Calculate Otsu's threshold
    threshold_value = threshold_otsu(image)
    
    # Apply Otsu's thresholding to create binary image
    binary_image = image > threshold_value
    
    # Display the binary image
    ax[i].imshow(binary_image, cmap='gray')
    ax[i].set_title(f'pic{i+1} Otsu')

# Show the plot
plt.show()
```
